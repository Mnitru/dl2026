```python
def f(x):
    return x * x
def df(x):
    return 2 * x
def gradient_descent(x0, lr, epochs=1000, eps=1e-6):
    x = x0
    print("iter\t x\t\t f(x)")
    for i in range(epochs):
        fx = f(x)
        print(f"{i}\t {x:.6f}\t {fx:.6f}")
        grad = df(x)
        new_x = x - lr * grad
        if abs(new_x - x) < eps:
            x = new_x
            break
        x = new_x
    print(f"Minimum at x = {x:.6f}, f(x) = {f(x):.6f}")
    return x
gradient_descent(x0=5, lr=0.1)
```
